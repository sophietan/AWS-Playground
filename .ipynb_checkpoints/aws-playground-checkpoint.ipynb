```python
def get_spot_price(client, instanceType: list, region = 'eu-west-2', productDescription = ['Linux/UNIX']):
    zones = [region + zone for zone in ['a', 'b', 'c']]
    prices = []
    for zone in zones:
        response = ec2.describe_spot_price_history(AvailabilityZone = zone, 
                                                   ProductDescriptions = productDescription, 
                                                   MaxResults = 100,
                                                   InstanceTypes = instanceType
                                                  )
        prices.extend(response['SpotPriceHistory'])
    return pd.DataFrame(prices)
```

```python
df = get_spot_price(ec2, ['i3.4xlarge'])
```
